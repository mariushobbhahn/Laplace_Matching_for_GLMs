# Exponential Distribution

```python
def exponential_pdf(x, l):
    return(l*np.exp(-l*x))

def exponential_pdf_transform(x, l):
    return(l*np.exp(-l*np.exp(x) + x))


def exponential_mu_trans(l):
    return(np.log(1/l))

def exponential_sigma_trans(l):
    return(1)

def exponential_normal_back(x, mu, sigma):
    x_ = np.log(x)
    return(norm.pdf(x_, mu, sigma) * 1/x)


x = np.linspace(0,5,1000+1)
x_trans = np.linspace(-10,5,1000+1)

l1 = 1
l2 = 2
l3 = 3

##### original
y1 = [exponential_pdf(x_, l1) for x_ in x]
y2 = [exponential_pdf(x_, l2) for x_ in x]
y3 = [exponential_pdf(x_, l3) for x_ in x]


##### transformation
y1_trans = [exponential_pdf_transform(x_, l1) for x_ in x_trans]
y2_trans = [exponential_pdf_transform(x_, l2) for x_ in x_trans]
y3_trans = [exponential_pdf_transform(x_, l3) for x_ in x_trans]

mu1_trans = exponential_mu_trans(l1)
mu2_trans = exponential_mu_trans(l2)
mu3_trans = exponential_mu_trans(l3)

sigma1_trans = 1
sigma2_trans = 1
sigma3_trans = 1

y1_trans_lpa = norm.pdf(x_trans, mu1_trans, sigma1_trans)
y2_trans_lpa = norm.pdf(x_trans, mu2_trans, sigma2_trans)
y3_trans_lpa = norm.pdf(x_trans, mu3_trans, sigma3_trans)

##### back transform
y1_back = exponential_normal_back(x, mu1_trans, sigma1_trans)
y2_back = exponential_normal_back(x, mu2_trans, sigma2_trans)
y3_back = exponential_normal_back(x, mu3_trans, sigma3_trans)

plt.rcParams.update({'font.size': 20})
fig, axs = plt.subplots(nrows=1, ncols=3, sharex=False, figsize=(26,6))
axs[0].plot(x, y1, 'r-', label="exponential");
axs[0].plot(x, y2, 'b-')#, label="exponential");
axs[0].plot(x, y3, 'k-')#, label="exponential");
axs[0].legend();

axs[1].plot(x_trans, y1_trans, 'r-', label="exponential");
axs[1].plot(x_trans, y2_trans, 'b-')#, label="exponential");
axs[1].plot(x_trans, y3_trans, 'k-')#, label="exponential");
axs[1].plot(x_trans, y1_trans_lpa, 'r--', label="Laplace approximation");
axs[1].plot(x_trans, y2_trans_lpa, 'b--')#, label="LPA of chi2_2");
axs[1].plot(x_trans, y3_trans_lpa, 'k--')#, label="LPA of chi2_3");
axs[1].legend();

axs[2].plot(x, y1, 'r-', label="exponential");
axs[2].plot(x, y2, 'b-')#, label="exponential");
axs[2].plot(x, y3, 'k-')#, label="exponential");
axs[2].plot(x, y1_back, 'r--', label="back transform")
axs[2].plot(x, y2_back, 'b--')#, label="back transform")
axs[2].plot(x, y3_back, 'k--')#, label="back transform")
axs[2].legend();

plt.savefig('exponential_playground.pdf')
```

```python
########## Square basis transform ###########

def exponential_pdf(x, l):
    return(l*np.exp(-l*x))

def exponential_pdf_transform_sqrt(x, l):
    return(l*np.exp(-l*x**2 + np.log(x)))


def exponential_mu_trans_sqrt(l):
    return(np.sqrt(1/(2*l)))

def exponential_sigma_trans_sqrt(l):
    return(1/(4*l))

def exponential_normal_back(x, mu, sigma):
    x_ = np.log(x)
    return(norm.pdf(x_, mu, sigma) * 1/x)


x = np.linspace(0,5,1000+1)
x_trans = np.linspace(-1,3,1000+1)

l1 = 1
l2 = 2
l3 = 3

##### original
y1 = [exponential_pdf(x_, l1) for x_ in x]
y2 = [exponential_pdf(x_, l2) for x_ in x]
y3 = [exponential_pdf(x_, l3) for x_ in x]


##### transformation
y1_trans = [exponential_pdf_transform_sqrt(x_, l1) for x_ in x_trans]
y2_trans = [exponential_pdf_transform_sqrt(x_, l2) for x_ in x_trans]
y3_trans = [exponential_pdf_transform_sqrt(x_, l3) for x_ in x_trans]

mu1_trans = exponential_mu_trans_sqrt(l1)
mu2_trans = exponential_mu_trans_sqrt(l2)
mu3_trans = exponential_mu_trans_sqrt(l3)

sigma1_trans = exponential_sigma_trans_sqrt(l1)
sigma2_trans = exponential_sigma_trans_sqrt(l2)
sigma3_trans = exponential_sigma_trans_sqrt(l3)

y1_trans_lpa = norm.pdf(x_trans, mu1_trans, sigma1_trans) * 0.2
y2_trans_lpa = norm.pdf(x_trans, mu2_trans, sigma2_trans) * 0.2
y3_trans_lpa = norm.pdf(x_trans, mu3_trans, sigma3_trans) * 0.2

##### back transform
y1_back = exponential_normal_back(x, mu1_trans, sigma1_trans)
y2_back = exponential_normal_back(x, mu2_trans, sigma2_trans)
y3_back = exponential_normal_back(x, mu3_trans, sigma3_trans)

plt.rcParams.update({'font.size': 20})
fig, axs = plt.subplots(nrows=1, ncols=3, sharex=False, figsize=(26,6))
axs[0].plot(x, y1, 'r-', label="exponential");
axs[0].plot(x, y2, 'b-')#, label="exponential");
axs[0].plot(x, y3, 'k-')#, label="exponential");
axs[0].legend();

axs[1].plot(x_trans, y1_trans, 'r-', label="exponential");
axs[1].plot(x_trans, y2_trans, 'b-')#, label="exponential");
axs[1].plot(x_trans, y3_trans, 'k-')#, label="exponential");
axs[1].plot(x_trans, y1_trans_lpa, 'r--', label="Laplace approximation");
axs[1].plot(x_trans, y2_trans_lpa, 'b--')#, label="LPA of chi2_2");
axs[1].plot(x_trans, y3_trans_lpa, 'k--')#, label="LPA of chi2_3");
axs[1].legend();

axs[2].plot(x, y1, 'r-', label="exponential");
axs[2].plot(x, y2, 'b-')#, label="exponential");
axs[2].plot(x, y3, 'k-')#, label="exponential");
axs[2].plot(x, y1_back, 'r--', label="back transform")
axs[2].plot(x, y2_back, 'b--')#, label="back transform")
axs[2].plot(x, y3_back, 'k--')#, label="back transform")
axs[2].legend();

plt.savefig('exponential_sqrt_playground.pdf')
```

# Gamma Distribution

```python
#change the basis, i.e. x -> t(x)

#######EXPONENTIAL BASIS TRANSFORM########
def gamma_func(n):
    return(np.math.factorial(n-1))

def gamma_pdf(x, a, l):
    r = l**a/gamma_func(a) * (x ** (a-1)) * (np.exp(-x*l))
    return(r)

def gamma_pdf_trans(x, a, l):
    r = l**a/gamma_func(a) * (np.exp(x) ** a) * (np.exp(-np.exp(x) * l))
    return(r)
    
"""Transforms"""

def square_transform(a):
    return(a**2)

def sqrt_transform(a):
    return(np.sqrt(a))

def exp_transform(a):
    return(np.exp(a))

def gamma_mode(a, l):
    return((a-1)/l)

def gamma_trans_mode(a, l):
    return(np.log(a/l))

def sigma_gamma(a,l):
    return((a-1)/l**2)

def sigma_gamma_trans(a):
    return(1/a)

"""values"""

a1, l1 = 5,1
a2, l2 = 4,2
a3, l3 = 8,3

#normal
mode1 = gamma_mode(a1, l1)
mode2 = gamma_mode(a2, l2)
mode3 = gamma_mode(a3, l3)

sigma1 = np.sqrt(sigma_gamma(a1, l1))
sigma2 = np.sqrt(sigma_gamma(a2, l2))
sigma3 = np.sqrt(sigma_gamma(a3, l3))

x = np.linspace(0, 12, 12000 + 1)
x_trans = np.linspace(-1, 4, 5000 + 1)
y1 = gamma_pdf(x, a1, l1)
y2 = gamma_pdf(x, a2, l2)
y3 = gamma_pdf(x, a3, l3)
lp_approx1 = norm.pdf(x, loc=mode1, scale=sigma1)
lp_approx2 = norm.pdf(x, loc=mode2, scale=sigma2)
lp_approx3 = norm.pdf(x, loc=mode3, scale=sigma3)

#transform

y1_trans = gamma_pdf_trans(x_trans, a1, l1)
y2_trans = gamma_pdf_trans(x_trans, a2, l2)
y3_trans = gamma_pdf_trans(x_trans, a3, l3)
mode1_trans = np.log(a1 / l1)
mode2_trans = np.log(a2 / l2)
mode3_trans = np.log(a3 / l3)

sigma1_trans = np.sqrt(sigma_gamma_trans(a1))
sigma2_trans = np.sqrt(sigma_gamma_trans(a2))
sigma3_trans = np.sqrt(sigma_gamma_trans(a3))
lp_approx1_trans = norm.pdf(x_trans, loc=mode1_trans, scale=sigma1_trans)
lp_approx2_trans = norm.pdf(x_trans, loc=mode2_trans, scale=sigma2_trans)
lp_approx3_trans = norm.pdf(x_trans, loc=mode3_trans, scale=sigma3_trans)

#back transform
y1_back = exponential_normal_back(x, mode1_trans, sigma1_trans)
y2_back = exponential_normal_back(x, mode2_trans, sigma2_trans)
y3_back = exponential_normal_back(x, mode3_trans, sigma3_trans)

"""Plots"""
fig, axs = plt.subplots(nrows=1, ncols=3, sharex=False, figsize=(23, 6))
#first plot
axs[0].plot(x, y1, 'r-', lw=1, alpha=1, label='gamma a1 l1 pdf')
axs[0].plot(x, y2, 'b-', lw=1, alpha=1)#, label='gamma a2 l2 pdf')
axs[0].plot(x, y3, 'k-', lw=1, alpha=1)#, label='gamma a3 l3 pdf')
axs[0].plot(x, lp_approx1, 'r--', lw=1, alpha=1, label='lp approx1 pdf')
axs[0].plot(x, lp_approx2, 'b--', lw=1, alpha=1)#, label='lp approx2 pdf')
axs[0].plot(x, lp_approx3, 'k--', lw=1, alpha=1)#, label='lp approx3 pdf')
axs[0].legend();

#second plot
axs[1].plot(x_trans, y1_trans, 'r-', lw=1, alpha=1, label='gamma trans')
axs[1].plot(x_trans, y2_trans, 'b-', lw=1, alpha=1)#, label='gamma trans a2 l2 pdf')
axs[1].plot(x_trans, y3_trans, 'k-', lw=1, alpha=1)#, label='gamma trans a3 l3 pdf')
axs[1].plot(x_trans, lp_approx1_trans, 'r--', lw=1, alpha=1, label='lp approx trans')
axs[1].plot(x_trans, lp_approx2_trans, 'b--', lw=1, alpha=1)#, label='lp approx2 trans pdf')
axs[1].plot(x_trans, lp_approx3_trans, 'k--', lw=1, alpha=1)#, label='lp approx3 trans pdf')
axs[1].legend();

axs[2].plot(x, y1, 'r-', label="gamma");
axs[2].plot(x, y2, 'b-')#, label="exponential");
axs[2].plot(x, y3, 'k-')#, label="exponential");
axs[2].plot(x, y1_back, 'r--', label="back transform")
axs[2].plot(x, y2_back, 'b--')#, label="back transform")
axs[2].plot(x, y3_back, 'k--')#, label="back transform")
axs[2].legend();

plt.savefig('gamma_playground_log.pdf')
```

# Inverse Gamma Distribution

```python
#change the basis, i.e. x -> t(x)
from scipy.special import gamma

def inv_gamma_pdf(x, a, l):
    r = l**a/gamma(a) * (x ** (-a-1)) * (np.exp(-l/x))
    return(r)

def inv_gamma_pdf_trans(x, a, l):
    r = l**a/gamma(a) * (np.exp(x) ** -a) * (np.exp(-l/np.exp(x)))
    return(r)
    
"""Transforms"""

def exp_transform(a):
    return(np.exp(a))

def inv_gamma_mode(a, l):
    return(l/(a+1))

def inv_gamma_trans_mode(a, l):
    return(np.log(l/a))

def inv_gamma_sigma(a,l):
    return(l**2/(a+1)**3)

def inv_gamma_sigma_trans(a):
    return(1/a)

"""values"""

a1, l1 = 0.5,0.5
a2, l2 = 2,1
a3, l3 = 3,1
mode1 = inv_gamma_mode(a1, l1)
mode2 = inv_gamma_mode(a2, l2)
mode3 = inv_gamma_mode(a3, l3)

sigma1 = np.sqrt(inv_gamma_sigma(a1, l1))
sigma2 = np.sqrt(inv_gamma_sigma(a2, l2))
sigma3 = np.sqrt(inv_gamma_sigma(a3, l3))

x = np.linspace(0, 4, 4000 + 1)
x_trans = np.linspace(-3, 5, 8000 + 1)
y1 = inv_gamma_pdf(x, a1, l1) *3.2
y2 = inv_gamma_pdf(x, a2, l2) *1.55
y3 = inv_gamma_pdf(x, a3, l3) *1.35
lp_approx1 = norm.pdf(x, loc=mode1, scale=sigma1)
lp_approx2 = norm.pdf(x, loc=mode2, scale=sigma2)
lp_approx3 = norm.pdf(x, loc=mode3, scale=sigma3)

y1_trans = inv_gamma_pdf_trans(x_trans, a1, l1) * 1.15
y2_trans = inv_gamma_pdf_trans(x_trans, a2, l2) * 1.05
y3_trans = inv_gamma_pdf_trans(x_trans, a3, l3) * 1.05
mode1_trans = inv_gamma_trans_mode(a1, l1)
mode2_trans = inv_gamma_trans_mode(a2, l2)
mode3_trans = inv_gamma_trans_mode(a3, l3)

sigma1_trans = np.sqrt(inv_gamma_sigma_trans(a1))
sigma2_trans = np.sqrt(inv_gamma_sigma_trans(a2))
sigma3_trans = np.sqrt(inv_gamma_sigma_trans(a3))
lp_approx1_trans = norm.pdf(x_trans, loc=mode1_trans, scale=sigma1_trans)
lp_approx2_trans = norm.pdf(x_trans, loc=mode2_trans, scale=sigma2_trans)
lp_approx3_trans = norm.pdf(x_trans, loc=mode3_trans, scale=sigma3_trans)

#back
y1_back = exponential_normal_back(x, mode1_trans, sigma1_trans) * 1.9
y2_back = exponential_normal_back(x, mode2_trans, sigma2_trans) * 1.5
y3_back = exponential_normal_back(x, mode3_trans, sigma3_trans) * 1.35


"""Plots"""
fig, axs = plt.subplots(nrows=1, ncols=3, sharex=False, figsize=(23, 6))
#first plot
axs[0].plot(x, y1, 'r-', lw=1, alpha=1, label='inverse gamma')
axs[0].plot(x, y2, 'b-', lw=1, alpha=1)#, label='gamma a2 l2 pdf')
axs[0].plot(x, y3, 'k-', lw=1, alpha=1)#, label='gamma a3 l3 pdf')
axs[0].plot(x, lp_approx1, 'r--', lw=1, alpha=1, label='Laplace approx')
axs[0].plot(x, lp_approx2, 'b--', lw=1, alpha=1)#, label='lp approx2 pdf')
axs[0].plot(x, lp_approx3, 'k--', lw=1, alpha=1)#, label='lp approx3 pdf')
axs[0].legend();

#second plot
axs[1].plot(x_trans, y1_trans, 'r-', lw=1, alpha=1, label='inverse gamma trans')
axs[1].plot(x_trans, y2_trans, 'b-', lw=1, alpha=1)#, label='gamma trans a2 l2 pdf')
axs[1].plot(x_trans, y3_trans, 'k-', lw=1, alpha=1)#, label='gamma trans a3 l3 pdf')
axs[1].plot(x_trans, lp_approx1_trans, 'r--', lw=1, alpha=1, label='laplace approx trans')
axs[1].plot(x_trans, lp_approx2_trans, 'b--', lw=1, alpha=1)#, label='lp approx2 trans pdf')
axs[1].plot(x_trans, lp_approx3_trans, 'k--', lw=1, alpha=1)#, label='lp approx3 trans pdf')
axs[1].legend();

axs[2].plot(x, y1, 'r-', label="inverse gamma");
axs[2].plot(x, y2, 'b-')#, label="exponential");
axs[2].plot(x, y3, 'k-')#, label="exponential");
axs[2].plot(x, y1_back, 'r--', label="back transform")
axs[2].plot(x, y2_back, 'b--')#, label="back transform")
axs[2].plot(x, y3_back, 'k--')#, label="back transform")
axs[2].legend();

plt.savefig('inverse_gamma_playground.pdf')
```

# 3D Dirichlet distribution

```python
# just copy pasted from this link: http://blog.bogatron.net/blog/2014/02/02/visualizing-dirichlet-distributions/
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as tri
import math
from functools import reduce
from math import gamma
from operator import mul
from scipy.special import loggamma

corners = np.array([[0, 0], [1, 0], [0.5, 0.75**0.5]])
triangle = tri.Triangulation(corners[:, 0], corners[:, 1])

refiner = tri.UniformTriRefiner(triangle)
trimesh = refiner.refine_triangulation(subdiv=4)


"""
plt.figure(figsize=(8, 4))
for (i, mesh) in enumerate((triangle, trimesh)):
    plt.subplot(1, 2, i+ 1)
    plt.triplot(mesh)
    plt.axis('off')
    plt.axis('equal')
"""

midpoints = [(corners[(i + 1) % 3] + corners[(i + 2) % 3]) / 2.0 \
             for i in range(3)]
def xy2bc(xy, tol=1.e-3):
    '''Converts 2D Cartesian coordinates to barycentric.'''
    s = [(corners[i] - midpoints[i]).dot(xy - midpoints[i]) / 0.75 \
         for i in range(3)]
    return np.clip(s, tol, 1.0 - tol)

def logit_transform(x):
    
    #x_D = x[:-1]
    #print("x_D: ", x_D, x[-1])
    #out = np.log(x_D/x[-1])
    out = np.log(x/(1-x))
    return(out)

def beta_function(alpha):
    return(np.exp(np.sum([loggamma(a_i) for a_i in alpha]) - loggamma(np.sum(alpha))))    

class Dirichlet(object):
    def __init__(self, alpha):
        self._alpha = np.array(alpha)
        self._coef = 1/beta_function(alpha)
        
    def pdf(self, x):
        '''Returns pdf value for `x`.'''
        return self._coef * reduce(mul, [xx ** (aa - 1)
                                         for (xx, aa)in zip(x, self._alpha)])
    
class Normal3D(object):
    
    def __init__(self, mu, Sigma):
        self.mu = np.array(mu)
        self.Sigma = np.linalg.inv(np.array(Sigma))
        self.const = 1 / (np.sqrt((2*np.pi)**3 * np.linalg.det(Sigma)))
        
    def pdf(self, x):
        """Calculate 3D Gaussian"""
        y = self.const * np.exp(-0.5 * (x - self.mu).T @ self.Sigma @ (x - self.mu))
        return(y)
    
class logitNormal3D(object):
    
    def __init__(self, mu, Sigma):
        self.mu = np.array(mu)
        self.Sigma = np.linalg.inv(np.array(Sigma))
        self.const = 1/ (np.sqrt(np.linalg.det(2 * np.pi * self.Sigma)))
        
    def pdf(self, x):
        part_one = self.const / np.prod(x*(1-x))
        logit = logit_transform(x)
        part_two = np.exp(-0.5*(logit - self.mu).T @ self.Sigma @ (logit - self.mu))
        y = part_one * part_two
        return(y)

def draw_pdf_contours(dist, nlevels=200, subdiv=5, **kwargs):

    refiner = tri.UniformTriRefiner(triangle)
    trimesh = refiner.refine_triangulation(subdiv=subdiv)
    pvals = [dist.pdf(xy2bc(xy)) for xy in zip(trimesh.x, trimesh.y)]
    xys = [xy2bc(xy) for xy in zip(trimesh.x, trimesh.y)]
    #print(len(pvals))
    #print(xys)
    print(len(xys))

    plt.tricontourf(trimesh, pvals, nlevels, **kwargs)
    plt.axis('equal')
    plt.xlim(0, 1)
    plt.ylim(0, 0.75**0.5)
    plt.axis('off')
    
draw_pdf_contours(Dirichlet([5, 5, 5]))
```

```python
from scipy.stats import dirichlet, multivariate_normal

def argmin_norm(sample, positions):
    return(np.argmin(np.sum(np.square(sample - positions), axis=1)))

def sample_dirichlet_contours(alpha, nlevels=200, subdiv=5, num_samples=10000, **kwargs):

    refiner = tri.UniformTriRefiner(triangle)
    trimesh = refiner.refine_triangulation(subdiv=subdiv)
    xys = np.array([xy2bc(xy) for xy in zip(trimesh.x, trimesh.y)])
    dirichlet_samples = dirichlet.rvs(alpha, size=num_samples)
    counts = np.zeros(len(xys))
    for x in dirichlet_samples:
        counts[argmin_norm(x, xys)] += 1
    
    pvals = counts/num_samples

    plt.tricontourf(trimesh, pvals, nlevels, **kwargs)
    plt.axis('equal')
    plt.xlim(0, 1)
    plt.ylim(0, 0.75**0.5)
    plt.axis('off')
    plt.show()
    
def sample_Normal_contours(mu, sigma, nlevels=200, subdiv=5, num_samples=10000, **kwargs):

    refiner = tri.UniformTriRefiner(triangle)
    trimesh = refiner.refine_triangulation(subdiv=subdiv)
    xys = np.array([xy2bc(xy) for xy in zip(trimesh.x, trimesh.y)])
    gaussian_samples = multivariate_normal.rvs(mean=mu, cov=sigma, size=num_samples)
    softmax_gaussian_samples = np.array([softmax_transform(x) for x in gaussian_samples])
    counts = np.zeros(len(xys))
    for x in softmax_gaussian_samples:
        counts[argmin_norm(x, xys)] += 1
    
    pvals = counts/num_samples

    plt.tricontourf(trimesh, pvals, nlevels, **kwargs)
    plt.axis('equal')
    plt.xlim(0, 1)
    plt.ylim(0, 0.75**0.5)
    plt.axis('off')
    plt.show()
    
sample_dirichlet_contours(np.array([5, 5, 5]))
sample_Normal_contours(np.array([-1, 2, -1]), 0.1*np.eye(3))
```

```python
#change the basis, i.e. x -> t(x)
from scipy.special import gamma
from scipy.stats import norm

"""Transforms"""
def logit_transform(theta):
    return(np.log(theta/(1-theta)))

def logit_derivative(x):
    return(1/x + 1/(1-x))

#transformed to softmax space
def softmax_transform(a):
    x = np.exp(a)/np.sum(np.exp(a))
    return(x)

def logistic_transform(a):
    return(1 / (1 + np.exp(-a)))

def logistic_derivative(x):
    return(logistic_transform(x) * (1- logistic_transform(x)))

def B(a,b):
    nom = gamma(a) * gamma(b)
    denom = gamma(a + b)
    return(nom/denom)

def beta_pdf(x, a, b):
    nom = x ** (a-1) * (1 - x) ** (b-1)
    denom = B(a, b)
    return(nom/denom)

def beta_pdf_trans_logistic(x, a, b):
    x_t = logistic_transform(x)
    nom = (x_t) ** (a) * (1- x_t) ** (b)
    denom = B(a, b)
    return((nom/denom))

def beta_mode(a, b):
    return((a-1)/(a+b-2))

def beta_sigma(a, b):
    return(((a-1)*(b-1))/(a+b-2)**3)

def mode_beta_trans(a,b):
    return(np.log(a/b))

def beta_sigma_trans(a,b):
    return((a+b)/(a*b))

def alpha_beta_from_normal(mu, sigma):
    a = (np.exp(mu) + 1)/sigma**2
    b = (np.exp(-mu) + 1)/sigma**2

    return(a, b)

def normal_to_standard_basis(x, mu, sigma):
    x_ = logit_transform(x)
    return(norm.pdf(x_, mu, sigma) * logit_derivative(x))

"""values"""

a1, b1 = 0.8, 0.9
a2, b2 = 4, 2
a3, b3 = 2, 7
#a1, b1 = 2, 1.2
#a2, b2 = 0.5, 0.9
#a3, b3 = 3,4

mode1 = beta_mode(a1, b1)
mode2 = beta_mode(a2, b2)
mode3 = beta_mode(a3, b3)
print("modes: ", mode1, mode2, mode3)

#plot1: normal base
sigma1 = np.sqrt(beta_sigma(a1, b1))
sigma2 = np.sqrt(beta_sigma(a2, b2))
sigma3 = np.sqrt(beta_sigma(a3, b3))

x = np.linspace(0,1, 1000 + 1)
x2 = np.linspace(-3,7, 1000 + 1)
y1 = beta_pdf(x, a1, b1)
y2 = beta_pdf(x, a2, b2)
y3 = beta_pdf(x, a3, b3)

LP_approx1 = norm.pdf(x, mode1, sigma1)
LP_approx2 = norm.pdf(x, mode2, sigma2)
LP_approx3 = norm.pdf(x, mode3, sigma3)

#plot 2: transform to other base
y1_logistic = beta_pdf_trans_logistic(x2, a1, b1)
y2_logistic = beta_pdf_trans_logistic(x2, a2, b2)
y3_logistic = beta_pdf_trans_logistic(x2, a3, b3)

mode1_trans = mode_beta_trans(a1,b1)
mode2_trans = mode_beta_trans(a2,b2)
mode3_trans = mode_beta_trans(a3,b3)

sigma1_trans = np.sqrt(beta_sigma_trans(a1, b1))
sigma2_trans = np.sqrt(beta_sigma_trans(a2, b2))
sigma3_trans = np.sqrt(beta_sigma_trans(a3, b3))


LP_approx1_trans = norm.pdf(x2, mode1_trans, sigma1_trans)
LP_approx2_trans = norm.pdf(x2, mode2_trans, sigma2_trans)
LP_approx3_trans = norm.pdf(x2, mode3_trans, sigma3_trans)

#plot 3: transform back and reconstruct
#a1_rec, b1_rec = alpha_beta_from_normal(mode1_trans, sigma1_trans)
#a2_rec, b2_rec = alpha_beta_from_normal(mode2_trans, sigma2_trans)
#a3_rec, b3_rec = alpha_beta_from_normal(mode3_trans, sigma3_trans)

y1_rec = normal_to_standard_basis(x, mode1_trans, sigma1_trans)
y2_rec = normal_to_standard_basis(x, mode2_trans, sigma2_trans)
y3_rec = normal_to_standard_basis(x, mode3_trans, sigma3_trans)

"""Plots"""
fig, axs = plt.subplots(nrows=1, ncols=3, sharex=False, figsize=(23, 6))
#plot beta distribution and LPA in the standard basis
axs[0].plot(x, y1, 'r-', lw=1, alpha=1)#, label='beta 1 pdf')
axs[0].plot(x, y2, 'b-', lw=1, alpha=1)#, label='beta 2 pdf')
axs[0].plot(x, y3, 'k-', lw=1, alpha=1, label='1D Dirichlet')
axs[0].plot(x, LP_approx1, 'r--', lw=1, alpha=1)#, label='LP approx beta1')
axs[0].plot(x, LP_approx2, 'b--', lw=1, alpha=1)#, label='LP approx beta2')
axs[0].plot(x, LP_approx3, 'k--', lw=1, alpha=1, label='Laplace')
axs[0].set_yticklabels([])
axs[0].set_yticks([])
axs[0].set_xlabel('$\pi$')
axs[0].set_ylabel('$p(\pi)d\pi$')
axs[0].set_title('Laplace in standard basis')
axs[0].legend();

#plot beta distribution in the transformed basis
axs[1].plot(x2, y1_logistic, 'r-', lw=1, alpha=1)#, label='beta1 transformed pdf logistic')
axs[1].plot(x2, y2_logistic, 'b-', lw=1, alpha=1)#, label='beta2 transformed pdf logistic')
axs[1].plot(x2, y3_logistic, 'k-', lw=1, alpha=1, label='1D Dirichlet')
axs[1].plot(x2, LP_approx1_trans, 'r--', lw=1, alpha=1)#, label='LP approx beta1 trans')
axs[1].plot(x2, LP_approx2_trans, 'b--', lw=1, alpha=1)#, label='LP approx beta2 trans')
axs[1].plot(x2, LP_approx3_trans, 'k--', lw=1, alpha=1, label='Laplace')
axs[1].set_xlabel('x')
axs[1].set_ylabel('$p(x)dx$')
axs[1].set_yticklabels([])
axs[1].set_yticks([])
axs[1].set_title('Laplace in softmax basis')
axs[1].legend();

#plot backwards transformation of the LPA from the transformed to the standard basis
axs[2].plot(x, y1, 'r-', lw=1, alpha=1)#, label='beta 1 pdf')
axs[2].plot(x, y2, 'b-', lw=1, alpha=1)#, label='beta 2 pdf')
axs[2].plot(x, y3, 'k-', lw=1, alpha=1, label='1D Dirichlet')
axs[2].plot(x, y1_rec, 'r--', lw=1, alpha=1)#, label='beta 1 pdf')
axs[2].plot(x, y2_rec, 'b--', lw=1, alpha=1)#, label='beta 2 pdf')
axs[2].plot(x, y3_rec, 'k--', lw=1, alpha=1, label='Back transformation')
axs[2].set_yticklabels([])
axs[2].set_yticks([])
axs[2].set_xlabel('$\pi$')
axs[2].set_ylabel('$p(\pi)d\pi$')
axs[2].set_title('Transformation back to standard basis')
axs[2].legend();

plt.rcParams.update({'font.size': 20})
#plt.tight_layout()
#plt.show()
#plt.savefig('BetaVizTransformation.pdf')
#tpl.save('BetaVizTransformation.tex', figure=fig, tex_relative_path_to_data='.figures', figurewidth='\\figwidth', figureheight='\\figheight',
#         extra_axis_parameters={'xtick align=inside', 'ytick align=inside', 'xtick pos=left', 'ytick pos=left', 'legend pos=north east'})

plt.savefig('beta_playground.pdf')
```

```python
def my_kron_product(A, B):
    m1, n1 = A.shape[0], A.shape[1]
    m2, n2 = B.shape[0], B.shape[1]
    K = np.zeros((m1*m2, n1*n2))
    for i in range(m1):
        for j in range(m2):
            for k in range(n1):
                for l in range(n2):
                    K[i*m2 +j, k*n2 + l] = A[i,k] * B[j,l]
    
    return(K)
    
def my_box_product(A, B):
    m1, n1 = A.shape[0], A.shape[1]
    m2, n2 = B.shape[0], B.shape[1]
    K = np.zeros((m1*m2, n1*n2))
    for i in range(m1):
        for j in range(m2):
            for k in range(n1):
                for l in range(n2):
                    K[i*m2 +j, k*n1 + l] = A[i,l] * B[j,k]
    
    return(K)
```

```python
def inverse_wishart_trans_sigma_3(nu, p, Psi):
    
    X = inverse_wishart_trans_mode(nu, p, Psi)
    X_inv = np.linalg.inv(X)
    X3_inv = X_inv @ X_inv @ X_inv
    X_inv = np.linalg.inv(X)
    X2 = X@X
    
    part_one = (nu+p) * my_box_product(X_inv.T, X_inv)
    
    n = Psi.shape[0]
    K = np.zeros((n**2, n**2))
    for k in range(n):
        for i in range(n):
            for j in range(n):
                for l in range(n):
                    x1 = X[i,l] if k == j else 0
                    K[n*k + j, n*i + l] = x1

    K2 = -np.kron(X3_inv, X3_inv.T)
    K3 = np.kron(np.eye(2), X2.T) + np.kron(X, X.T) + np.kron(X2, np.eye(2))
    
    final = -(part_one + (K @ K2 @ K3).T)
    final_inv = np.linalg.inv(final)
    return(final_inv)
```
