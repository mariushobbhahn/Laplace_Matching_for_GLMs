# Gamma log bridge

```python
def gamma_mu(a, l):
    return(np.log(a/l))

def gamma_sigma2(a):
    return(1/a)

def gamma_alpha(sigma):
    return(1/sigma**2)

def gamma_lambda(mu, sigma):
    return(1/(np.exp(mu)*sigma**2))
```

```python
#verify the bridge

test_a, test_l = 3,4
test_m, test_s = gamma_mu(test_a, test_l), np.sqrt(gamma_sigma2(test_a))
test_a2, test_l2 = gamma_alpha(test_s), gamma_lambda(test_m, test_s)

print(test_a2, test_l2)

rand_a, rand_l = np.random.random(), np.random.random()
rand_m, rand_s = gamma_mu(rand_a, rand_l), np.sqrt(gamma_sigma2(rand_a))
rand_a2, rand_l2 = gamma_alpha(rand_s), gamma_lambda(rand_m, rand_s)

assert(np.allclose(rand_a, rand_a2))
assert(np.allclose(rand_l2, rand_l2))
```

# Beta logit bridge

```python
def beta_mu(a, b):
    return(np.log(a/b))

def beta_sigma2(a,b):
    return((a+b)/(a*b))

def beta_alpha(mu, sigma):
    return((np.exp(mu) + 1)/sigma**2)

def beta_beta(mu, sigma):
    return((np.exp(-mu) + 1)/sigma**2)
```

```python
#verify the bridge

test_a, test_b = 3,4
test_m, test_s = beta_mu(test_a, test_b), np.sqrt(beta_sigma2(test_a, test_b))
test_a2, test_b2 = beta_alpha(test_m, test_s), beta_beta(test_m, test_s)

print(test_a2, test_b2)

rand_a, rand_b = np.random.random(), np.random.random()
rand_m, rand_s = beta_mu(rand_a, rand_b), np.sqrt(beta_sigma2(rand_a, rand_b))
rand_a2, rand_b2 = beta_alpha(rand_m, rand_s), beta_beta(rand_m, rand_s)

assert(np.allclose(rand_a, rand_a2))
assert(np.allclose(rand_l2, rand_l2))
```

# Danger Zone

```python
mu_gamma = gamma_mu(a1, l1)
sigma2_gamma = gamma_sigma2(a1)

mu_beta = beta_mu(alpha1, beta1)
sigma2_beta = beta_sigma2(alpha1, beta1)

mu_combi = mu_beta + mu_gamma
sigma2_combi = sigma2_gamma + sigma2_beta

x = np.linspace(-5,5,1000+1)
```

```python
gamma_LP = norm.pdf(x, mu_gamma, np.sqrt(sigma2_gamma))
plt.plot(x, gamma_LP)
plt.show();
```

```python
beta_LP = norm.pdf(x, mu_beta, np.sqrt(sigma2_beta))
plt.plot(x, beta_LP)
plt.show();
```

```python
combi_LP = norm.pdf(x, mu_combi, np.sqrt(sigma2_combi))
plt.plot(x, combi_LP)
plt.show();
```

```python
def log_normal_back(x, mu, sigma):
    x_ = np.log(x)
    return(norm.pdf(x_, mu, sigma) * 1/x)

def logit_transform(theta):
    return(np.log(theta/(1-theta)))

def logit_derivative(x):
    return(1/x + 1/(1-x))

def normal_to_standard_basis(x, mu, sigma):
    x_ = logit_transform(x)
    return(norm.pdf(x_, mu, sigma) * logit_derivative(x))
```

```python
def test_back(x, mu, sigma):
    x_ = logit_transform(x) + np.log(x)
    return(norm.pdf(x_, mu, sigma)* (logit_derivative(x) + 1/x))
```
