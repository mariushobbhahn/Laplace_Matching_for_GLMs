```python
from scipy.special import gamma as gamma_func

def gamma_pdf(x, a, b):
    r = b**a/gamma_func(a) * (x ** (a-1)) * (np.exp(-x*b))
    return(r)
```

```python
def get_counts_and_num_awards(df):
    
    df_new = df.groupby(['math']).size().reset_index(name='counts')
    df_new['sum_awards'] = df.groupby(['math']).sum().reset_index()['num_awards']
    
    return(df_new)
```

```python
poisson_general_combined_df = get_counts_and_num_awards(poisson_general_df)
poisson_academic_combined_df = get_counts_and_num_awards(poisson_academic_df)
poisson_vocational_combined_df = get_counts_and_num_awards(poisson_vocational_df)
```

```python
def add_alpha_beta(df, alpha_0=0.1, beta_0=0.1):
    
    df['alpha'] = df['sum_awards'] + alpha_0
    df['beta'] = df['counts'] + beta_0
    return(df)
```

```python
poisson_general_combined_df = add_alpha_beta(poisson_general_combined_df)
poisson_academic_combined_df = add_alpha_beta(poisson_academic_combined_df)
poisson_vocational_combined_df = add_alpha_beta(poisson_vocational_combined_df)
```

```python
def add_mu_sigma(df):
    
    df['mu'] = np.log(df['alpha'] / df['beta'])
    df['sigma2'] = 1 / df['alpha']
    return(df)
```

```python
poisson_general_combined_df = add_mu_sigma(poisson_general_combined_df)
poisson_academic_combined_df = add_mu_sigma(poisson_academic_combined_df)
poisson_vocational_combined_df = add_mu_sigma(poisson_vocational_combined_df)
```

# Use GP+LM

```python
# starting with a GP for the general group

def k_SEK(x, x_, s=0.01, l=2):
    return(s*np.exp(-(x-x_)**2 / (2*l**2)))

def k_RQ(x, x_, s=0.01, l=2, a=0.1):
    return(s**2 * (1 + (x-x_)**2/(2*a*l**2)))**(-a)

X_general = poisson_general_combined_df['math']
n_general = len(X_general)

K_XX_general = np.zeros((n_general, n_general))
for i in range(n_general):
    for j in range(n_general):
        m_i = X_general[i]       
        m_j = X_general[j]
        K_XX_general[i, j] = k_SEK(m_i, m_j)
        #K_XX_general[i, j] = k_RQ(m_i, m_j)
        
K_XX_general += np.eye(n_general) * 1e-6
        
print(np.shape(K_XX_general))
K_XX_general[:10,:10]

plt.imshow(K_XX_general)
plt.show()
```

```python
# prep y data

Y_general = poisson_general_combined_df['mu']

# prep sigmas as noise for Kernel

# TODO
```

```python
# compute the equations for the GP

from numpy.random import randn, multivariate_normal
from numpy.linalg import cholesky, solve
from scipy.linalg import cho_solve, cho_factor
import time

G_ = K_XX_general #+ Y_Sigma_block
t0 = time.time()
G = cho_factor(G_)
t1 = time.time()
print("time for cho factor: ", t1 - t0)
```

```python
#make predictions for all math scores in between the existing ones and below or above the last known score

x_general = np.array(range(X_general.min() - 10, X_general.max() + 10))

print(np.shape(x_general))

n_general_ = len(x_general)
K_xX_general = np.zeros((n_general_, n_general))
for i in range(n_general_):
    for j in range(n_general):
        m_i = x_general[i]       
        m_j = X_general[j]
        K_xX_general[i, j] = k_SEK(m_i, m_j)
        #K_xX_general[i, j] = k_RQ(m_i, m_j)
        
print(np.shape(K_xX_general))
plt.imshow(K_xX_general)
plt.show();

K_xx_general = np.zeros((n_general_, n_general_))
for i in range(n_general_):
    for j in range(n_general_):
        m_i = x_general[i]       
        m_j = x_general[j]
        K_xx_general[i, j] = k_SEK(m_i, m_j)
        #K_xx_general[i, j] = k_RQ(m_i, m_j)
        
print(np.shape(K_xx_general))
plt.imshow(K_xx_general)
plt.show();
```

```python
# do some inference 
t0 = time.time()
A = cho_solve(G, K_xX_general.T).T
t1 = time.time()
print("cho solve: ", t1-t0)

out = {}

# # posterior p(f_x|Y) = N(f_x,mpost,vpost)
t0 = time.time()
out["mpost"] = A @ Y_general #m(x) + A @ (Y - mX)  # mean
t1 = time.time()
print("matrix multiplication with mu: ", t1-t0)
print(np.shape(out['mpost']))

t0 = time.time()
out["vpost"] = K_xx_general - A @ K_xX_general.T  # covariance
t1 = time.time()
print("Computing Kernel for prediction: ", t1-t0)
print(np.shape(out['vpost']))

n_GP_samples = 50

#This makes the kernel die because it's too large
out["spost"] = multivariate_normal(
    out["mpost"].flatten(), out["vpost"], size=n_GP_samples
)
```

```python
general_std = np.sqrt(np.diag(K_xx_general))
general_mean = out["mpost"]
plt.plot(general_mean)
plt.plot(general_mean + 2*general_std)
plt.plot(general_mean - 2*general_std)
plt.show()
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import DotProduct, WhiteKernel, RBF, Matern, ExpSineSquared, RationalQuadratic

kernel = RationalQuadratic(length_scale=2) + DotProduct(sigma_0=4)
gpr = GaussianProcessRegressor(kernel=kernel,
        random_state=0, n_restarts_optimizer=10, optimizer=None,
        alpha=np.sqrt(poisson_general_combined_df['sigma2'].values)
                              ).fit(X_general.values.reshape(-1,1), Y_general)

general_mean2, general_std2 = gpr.predict(x_general.reshape(-1,1), return_std=True)

plt.figure(figsize=(15, 7))

plt.plot(poisson_general_df['math'], poisson_general_df['num_awards']+eps1, '.', label="general", color='orange')
plt.plot(x_general, np.exp(general_mean2))
plt.plot(x_general, np.exp(general_mean2 + 2*general_std2))
plt.plot(x_general, np.exp(general_mean2 - 2*general_std2))
plt.plot();
```
