# Data Wrangling

We will extract the "Wahlbezirke" for Tübingen of all Federal elections between 1983 and 2017. This is done through the following process.
- find the correct identifier for Tübingen and select all rows that belong to this identifier
- create a new dataframe which includes the following parties: "CDU/CSU", "SPD", "FDP", "GRÜNE", "LINKE", "AFD", "PIRATEN", "DIE PARTEI" and "OTHER". 
- This means that every df has the following columns (in that order): "Wahlbezirk", "Gültig", "CDU/CSU",  	"SPD", "FDP", "GRÜNE", "DIE LINKE", "AfD", "PIRATEN" ,"DIE PARTEI", "OTHER"
- TODO: select only those Wahlkreise that have existed since 1983 (If that is possible)

```python
# Data Wrangling for the election in 1983

# Tübingen has Wahlkreis identifier 194
# The town of Tübingen has Land 8, Regierungsbezirk 4, Kreis 16, and Gemeinde 41


#elections_1983_Bezirke_df.head()

elections_1983_Tübingen_df = elections_1983_Bezirke_df.loc[(elections_1983_Bezirke_df["Wahlkreis"] == 194) &
                              (elections_1983_Bezirke_df["Land"] == 8) &
                              (elections_1983_Bezirke_df["Regierungsbezirk"] == 4) &
                              (elections_1983_Bezirke_df["Kreis"] == 16) &
                              (elections_1983_Bezirke_df["Gemeinde"] == 41)&
                              (elections_1983_Bezirke_df["Bezirksart"] == 0)]

elections_1983_Tübingen_clean_df = elections_1983_Tübingen_df[[
    'Wahlbezirk', 'Gültige', 'CDU/CSU', 'SPD', 'FDP', 'GRÜNE'
]]


elections_1983_Tübingen_clean_df['DIE LINKE'] = np.zeros(len(elections_1983_Tübingen_clean_df))
elections_1983_Tübingen_clean_df['AfD'] = np.zeros(len(elections_1983_Tübingen_clean_df))
elections_1983_Tübingen_clean_df['PIRATEN'] = np.zeros(len(elections_1983_Tübingen_clean_df))
elections_1983_Tübingen_clean_df['DIE PARTEI'] = np.zeros(len(elections_1983_Tübingen_clean_df))

elections_1983_Tübingen_clean_df['OTHER'] = elections_1983_Tübingen_df['BWK']\
                                          + elections_1983_Tübingen_df['CBV']\
                                          + elections_1983_Tübingen_df['DKP']\
                                          + elections_1983_Tübingen_df['EAP']\
                                          + elections_1983_Tübingen_df['KPD']\
                                          + elections_1983_Tübingen_df['NPD']\
                                          + elections_1983_Tübingen_df['ÖDP']\
                                          + elections_1983_Tübingen_df['UST']\
                                          + elections_1983_Tübingen_df['Übrige']

#check that the votes add up
A = elections_1983_Tübingen_clean_df['Gültige'] - elections_1983_Tübingen_clean_df['CDU/CSU'] \
  - elections_1983_Tübingen_clean_df['SPD'] - elections_1983_Tübingen_clean_df['FDP'] \
  - elections_1983_Tübingen_clean_df['GRÜNE'] - elections_1983_Tübingen_clean_df['DIE LINKE'] \
  - elections_1983_Tübingen_clean_df['AfD'] - elections_1983_Tübingen_clean_df['PIRATEN'] \
  - elections_1983_Tübingen_clean_df['DIE PARTEI']
B = elections_1983_Tübingen_clean_df['OTHER']
assert(np.allclose(A, B))

#rename
elections_1983_Tübingen_clean_df.rename(columns={'Gültige': 'Gültig'}, inplace=True)

#reorder
elections_1983_Tübingen_clean_df = elections_1983_Tübingen_clean_df[
    ["Wahlbezirk", "Gültig", "CDU/CSU", "SPD", "FDP", "GRÜNE", 
         "DIE LINKE", "AfD", "PIRATEN" ,"DIE PARTEI", "OTHER"]
]

#save
elections_1983_Tübingen_clean_df.to_csv('Bundestagswahlen_Wahlbezirke_clean/Tübingen_1983.csv')

elections_1983_Tübingen_clean_df
```
